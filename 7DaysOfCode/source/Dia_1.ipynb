```python
# Importando bibliotecas
import pandas as pd
```

```python
# carragando os dados e os contatenando
dados_20101 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20101.csv')
dados_20102 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20102.csv')
dados_20111 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20111.csv')
dados_20112 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20112.csv')
dados_20121 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20121.csv')
dados_20122 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20122.csv')
dados_20131 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20131.csv')
dados_20132 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20132.csv')
dados_20141 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20141.csv')
dados_20142 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20142.csv')
dados_20151 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20151.csv')
dados_20152 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20152.csv')
dados_20161 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20161.csv')
dados_20162 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20162.csv')
dados_20171 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20171.csv')
dados_20172 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20172.csv')
dados_20181 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20181.csv')
dados_20182 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20182.csv')
dados_20191 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20191.csv')
dados_20192 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20192.csv')
dados_20201 = pd.read_csv('../Datasets/dados_emprestimos/emprestimos-20201.csv')

emprestimos = pd.concat([dados_20101, dados_20102, dados_20111, dados_20112, dados_20121, dados_20122, dados_20131, 
                        dados_20132, dados_20141, dados_20142, dados_20151, dados_20152, dados_20161, dados_20162, 
                        dados_20171, dados_20172, dados_20181, dados_20182, dados_20191, dados_20192, dados_20201])

emprestimos
```

```python
# Removendo duplicados
emprestimos = emprestimos.drop_duplicates()
# removendo valores nulos exceto para data_devolucao
emprestimos = emprestimos.dropna(subset=['id_emprestimo', 'codigo_barras', 'data_emprestimo', 'data_devolucao', 
                                         'matricula_ou_siape', 'tipo_vinculo_usuario', ])
emprestimos.head()
```

```python
#Importando dados de exemplares
dados_exemplares = pd.read_parquet('../Datasets/dados_exemplares.parquet')
dados_exemplares.head()
```

```python
#concatenando os dados
dados = pd.merge(emprestimos, dados_exemplares, on='codigo_barras')
dados.head()
```
